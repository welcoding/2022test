# 지역별 축제 알아보기

```python
import csv
data = csv.reader(open('culture.csv'))
next(data)

x = []
y = []
z = []
h = []

print('축제 알리미입니다.')
a = input('알고싶은 지역을 입력해주세요 : ')

for row in data : 
  if a in row[-1] :
     x.append(row[0])

print(x)

print('위 축제 중 가고 싶은 축제의 날짜를 알고싶으면 축제 이름을 입력해주세요')
b = input('가고 싶은 축제의 이름은 무엇인가요? : ')

import csv
data = csv.reader(open('culture.csv'))
next(data)

for row in data :
  if b in row[0] :
    y.append(row[2])
    z.append(row[4])
    h.append(row[-9])

print(str(y) +'축제가 시작됩니다.')
print('-------------------------------------------------')
print('축제의 내용은' + str(z) + '입니다.')
print('-------------------------------------------------')
print('자세한 내용을 알고 싶으시면 아래 주소로 접속해주세요.')
print(h)
```
